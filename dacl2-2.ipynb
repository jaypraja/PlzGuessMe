```python
import pandas as pd
import numpy as np
students = [[89,'M','verygood'],[95,'F','excellent'],[75,None,'good'],[np.NaN,'M','average'],[71,'M','good'],[np.NaN, None,'verygood'],[96,'F','verygood'],[98,'M','excellent']]

dfstudents = pd.DataFrame(students)
dfstudents.columns = ['Marks', 'Gender','Result']
dfstudents
```

```python
dfstudents.isnull().values.sum()
```

```python
import numpy as np

x = dfstudents.iloc[:, 0:2].values
y = dfstudents.iloc[:, 2].values

print(x)
print(y)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer_x = SimpleImputer(missing_values=np.NaN, strategy='median')
x[:, 0:1] = imputer_x.fit_transform(x[:, 0:1])

imputer_df = SimpleImputer(missing_values=None, strategy='most_frequent')
dfstudents['Gender'] = imputer_df.fit_transform(dfstudents['Gender'].values.reshape(-1, 1))[:, 0]

print("Corrected DataFrame:")
print(dfstudents)
```

```python
import pandas as pd
file_path = 'CountryAgeSalary.csv'
dfsal = pd.read_csv(file_path)
print("Original DataFrame:")
print(dfsal)

from sklearn.impute import SimpleImputer

imputer_age = SimpleImputer(missing_values=np.NaN, strategy='median')
dfsal['Age'] = imputer_age.fit_transform(dfsal['Age'].values.reshape(-1, 1))

imputer_salary = SimpleImputer(missing_values=np.NaN, strategy='median')
dfsal['Salary'] = imputer_salary.fit_transform(dfsal['Salary'].values.reshape(-1, 1))

print("\nCorrected DataFrame:")
print(dfsal)

```
